```python
import pandas as pd
import numpy as np
from IPython.display import display, HTML

from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import IsolationForest
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import kmapper as km

import matplotlib.pyplot as plt

# Data import
data, target = load_breast_cancer(as_frame=True, return_X_y=True)
data["target"] = pd.Categorical(target)
```

```python
# Data discovery

display(HTML("<strong>Dataset preview</strong>"))
display(data.head(5))

display(HTML("<strong>Dataset Statistics</strong>"))
display(data.describe())

display(HTML("<strong>Count benign samples vs malignant: 1 is for benign</strong>"))
display(pd.DataFrame.from_dict(data['target'].value_counts(ascending=True)))
```

```python
# Data discovery classes comparison
# de-skew dataset
malignant_df = data[data['target'] == 0]
benign_df = data[data['target'] == 1].head(len(malignant_df))

features = data.columns
features = features[:-1]

fig, axes = plt.subplots(nrows=10, ncols=3, figsize=(20, 30))
fig.subplots_adjust(hspace=0.5)

for i, (ax, feature) in enumerate(zip(axes.flatten(), features)):
	ax.hist(malignant_df[feature], bins=30, alpha=0.6, label='malignant')
	ax.hist(benign_df[feature], bins=30, alpha=0.6, label='benign')
	ax.legend(loc='upper right')
	ax.set_title(feature + ": maligant vs benign")
```

```python
# MAPPER application

# Create a 1-D lens with Isolation Forest
lens1_model = IsolationForest(random_state=42)
lens1_model.fit(data)
lens_if = lens1_model.decision_function(data).reshape((data.shape[0], 1))

# Create another 1-D lens with PCA
mapper = km.KeplerMapper(verbose=0)
lens_pca = mapper.fit_transform(data, projection=PCA(n_components=1))

colors = np.array(target).reshape(-1,1)
plt.scatter(lens_if, lens_pca, c=colors, alpha=0.5)
plt.title("Lenses data compaction")
plt.xlabel("Isolation Forest")
plt.ylabel("PCA")
plt.show()
```

```python
# Combine lenses pairwise to get a 2-D lens [Isolation Forest, PCA] lens
lenses = np.c_[lens_if, lens_pca]

# Create a simplicial complex from projected data
simplicial_complex = mapper.map(lenses,
                      data,
                      nr_cubes=20,
                      overlap_perc=0.6,
                      clusterer=KMeans(n_clusters=2, random_state=42))

mapper.visualize(simplicial_complex,
                 title="Breast cancer data simplicial complex",
                 path_html="mapper.html",
                 color_function=colors,
                 custom_tooltips=colors)

print("Saved to mapper.html")
```
